```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x + 3) ** 2

def df(x):
    return 2 * (x + 3)

x = 2 

alpha = 0.1

iteration = 27

x_values = [x]
y_values = [f(x)]

print("Iteration \t x\t\t f(x)")

for i in range(iteration):
    gradient = df(x)
    x = x - alpha * gradient
    x_values.append(x)
    y_values.append(f(x))
    print(f"iteration {i+1}: x = {x}, f(x) = {f(x)}")

print("\n Local minima occurs near x =", round(x,4))

x_curve = np.linspace(-8, 5, 100)
y_curve = f(x_curve)
```

```python
plt.plot(x_curve, y_curve, label='f(x) = (x + 3)^2')

plt.scatter(x_values, y_values, label='Gradient Descent Steps')
plt.plot(x_values, y_values)

plt.title('Gradient Descent for finding local minimum')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```
